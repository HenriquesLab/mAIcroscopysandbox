```python
import math
from matplotlib import pyplot as plt
from maicroscopy_sandbox import mAIcroscopySandbox
from maicroscopy_sandbox.samples.binary import BinarySample

microscope = mAIcroscopySandbox(stage_size=[500, 500], fov_size=[500, 500])
sample = BinarySample(
    img_path="../binary_maps/sPiRaL.tif",
    sample_size=microscope.stage_size,
    bleaching_rate=0.00,
    movement_rate=5,
    movement_probability=0.0
)
microscope.set_laser_power(100)
frame = microscope.load_sample(sample, acquire=True)
```

```python
import numpy as np
from tifffile import imwrite
out = np.zeros((100, 500, 500))
for i in range(100):
    out[i] = microscope.acquire_image()
imwrite("sPiRaL_fluor.tif", out)
```

```python
from nanopyx import eSRRF

esrrf_out = eSRRF(out[0:200, :, :].astype(np.float32))
imwrite("sPiRaL_eSRRF_out.tif", esrrf_out)
```
